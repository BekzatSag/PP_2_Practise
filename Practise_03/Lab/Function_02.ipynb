```python
def categories (movies, category):
    flag = False
    list_mov = []
    for i in movies:
        if i["category"] == category:
            if not flag: flag = True
            list_mov.append(i["name"])
    if not flag: return "There is no such category"
    return list_mov
```

```python
def average_rating_names(movies, list_names):
    sum=0
    flag = False
    for i in movies:
        for j in list_names:
            if j==i["name"]:
                if not flag: flag = True
                sum+=i["imdb"]
        if not flag: return "There is no movie with some of the name"
        else: flag==False
    return sum/len(list_names)
```

```python
def average_rating_category(movies, category):
    sum = 0
    counter=0
    for i in movies:
        if i["category"] == category: 
            sum+=i["imdb"]
            counter+=1
    if counter == 0: return "There is no such category"
    return sum/counter
```
